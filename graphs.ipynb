```python
def graph(file_path, save_path):
    data = pd.read_csv(file_path)
    plt.style.use('ggplot')
    fig, ax = plt.subplots(nrows=1, ncols=2, figsize=(20, 8))

    ax[0].plot(data['Epoch'], data['Train Loss'], label='Training Loss', color='red')
    ax[0].plot(data['Epoch'], data['Eval Loss'], label='Validation Loss', color='blue')
    ax[0].set_title('Model Loss Curve')
    ax[0].set_xlabel('Epoch')
    ax[0].set_ylabel('Loss')
    ax[0].legend()


    ax[1].plot(data['Epoch'], data['Train Accuracy'], label='Training Accuracy', color='red')
    ax[1].plot(data['Epoch'], data['Top-1 Accuracy'], label='Top-1 Accuracy', color='blue')
    ax[1].plot(data['Epoch'], data['Top-5 Accuracy'], label='Top-5 Accuracy', color='green')
    ax[1].set_title('Model Accuracy Curve')
    ax[1].set_xlabel('Epoch')
    ax[1].set_ylabel('Accuracy')
    ax[1].legend()


    plt.tight_layout()
    plt.savefig(save_path)
    plt.close(fig)
```

```python
paths = [RESNET18, RESNET18_DROP, MOBILENETV2, MOBILENETV2_DROP, VGG11_DROP, VGG11_DROPDROP, DENSENET121, DENSENET121_DROP]
saves = [S_RESNET18, S_RESNET18_DROP,S_MOBILENETV2, S_MOBILENETV2_DROP, S_VGG11_DROP, S_VGG11_DROPDROP, S_DENSENET121, S_DENSENET121_DROP]

for file_path, save_path in zip(paths, saves):
    graph(file_path, save_path)
```
